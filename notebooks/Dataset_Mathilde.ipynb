```python
#Bref on va crér corpus CSV nous-mêmes !
from os import listdir
from textwrap import wrap
import numpy as np
import pandas as pd
import string

# Suppression des stop-words à partir de ma propre base de stop-words
with open('StopwordsLatin.txt', 'r') as stop_words_file:
    stop_words = stop_words_file.read().splitlines()
#Je liste tous les fichiers CSV dans mon répertoire courant
def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]
Liste_CSV=find_csv_filenames("./")
#J'affiche ma liste
Liste_CSV
#Attention, fais bien gaffe à n'avoir que des CSV du genre De_Pallio.csv
#Si tu as autres choses, crée un répertoire, par exemple Dataset et tu remplaces l'avant dernière ligne par 
#Liste_CSV=find_csv_filenames("./Dataset/")
#Il va lister les noms de fichiers CSV là-bas
```

```python
#Pour chaque texte, je fais 10 séquences, tu peux changer évidemment 
Nb_sequences=10
#Pour chaque CSV observé
for csv in Liste_CSV:
    #Je l'ouvre avec Pandas (=librairie ultra classique CF les bases que de mon cours, c'est à retenir)
    CSV=pd.read_csv(csv, sep=';')
    #Je crée ma longue chaîne de caractère
    text_csv=''
    #J'enlève les trous
    CSV['lemma']=CSV['lemma'].dropna()
    #Je compte les lignes qui ne sont pas des strings
    Count=0
    #Pour chaque ligne
    for word in CSV['lemma']:
        #Si c'est un string
        if type(word)==str:
            #Minuscule
            word=word.lower()
            #Si ce n'est pas un stop-words
            if word not in stop_words:
                #Je l'joute à ma chaîne de caractère
                text_csv+=word + ' '
    #J'enlève la ponctuation
    text_csv = text_csv.translate(str.maketrans('', '', string.punctuation))
    print("Nombres de lignes qui ne sont pas des string : " + str(Count))
    #Je coupe ma chaîne de caractère en 10
    phrases=wrap(text_csv, len(text_csv)/Nb_sequences)
    #Je crée une liste qui viendra indiquer l'auteur de la séquence en question
    auteur=np.array([csv[:-4]] *(Nb_sequences+1))
    d = {'text': phrases, 'author': auteur}
    #J'en fais un dataframe Pandas
    new_dataframe=pd.DataFrame(d)
    #Je l'enregistre dans ./Phrases/
    #Si tu as un problèmes, crée le répertoire à la main et appelles le "Phrases" donc
    new_dataframe.to_csv('./Phrases/' + csv[:-4]+ "_phrases.csv", sep=';')
```

## Créer Corpus.csv à partir des autres dans Phrases

```python
#Je crée une liste qui contient les noms des fichiers csv dans ./Phrases (=ceux générés avant)
Liste_CSV_phrases=find_csv_filenames("./Phrases/")
```

```python
#Je crée une liste des dataframes (=les matrices) issus de ces fichiers
Liste_df=[]
for csv in Liste_CSV_phrases:
    Liste_df.append(pd.read_csv('./Phrases/' +csv,sep=';'))
```

```python
#J'en fais la jointure (=je les colle les uns en dessous des autres)
#Oui c'est une ligne en Python...
result = pd.concat(Liste_df)
#Voilà le résultat
result
```

```python
#J'enlève une colonne pour minimiser l'information
result=result.iloc[: , 1:]
result
```
